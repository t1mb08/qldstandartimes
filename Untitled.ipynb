```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("all_race_summaries.csv")

# Display the first 5 rows
print("First 5 rows of the dataset:")
print(df.head())

# Display information about the dataset
print("\nDataset Information:")
print(df.info())

# Get descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values in Each Column:")
print(df.isnull().sum())

# List column names
print("\nColumn Names:")
print(df.columns)

print(df)
```

```python
# Convert the 'event_date' column to datetime
df['event_date'] = pd.to_datetime(df['event_date'])

# Check if the conversion was successful
print("Converted event_date to datetime:")
print(df['event_date'])

# Filter for dates between August 2023 and August 2024
start_date = pd.Timestamp('2023-08-01')
end_date = pd.Timestamp('2024-08-31')
filtered_df = df[(df['event_date'] >= start_date) & (df['event_date'] <= end_date)]

# Display the filtered DataFrame before dropping duplicates
print("\nFiltered DataFrame before dropping duplicates:")
print(filtered_df)

# Filter race names to keep only one race per row
df_filtered = filtered_df.drop_duplicates(subset=['race_name'])

# Round total_distance to the nearest 100
df_filtered['total_distance'] = df_filtered['total_distance'].round(-2)

# Display the final filtered DataFrame
print("\nFinal Filtered DataFrame from August 2023 to August 2024:")

# List of course names to filter out
courses_to_exclude = ['Rockhampton', 'Mackay', 'Townsville', 'Ladbrokes Cannon Park']

# Filter out specified course names
df_filtered = df_filtered[~df_filtered['course_name'].isin(courses_to_exclude)]
```

```python
# List of course names to filter out
courses_to_exclude = ['Rockhampton', 'Mackay', 'Townsville', 'Ladbrokes Cannon Park']

# Filter out specified course names
df_filtered = df[~df['course_name'].isin(courses_to_exclude)]

# Filter out specified course names
df_filtered = df_filtered[df_filtered['result_state'] != "DidNotStart"]

# Filter out specified course names

# Filtering races that contain "BENCHMARK" followed by a number or "Class" followed by a number
BnC = df_filtered[
    df_filtered['race_name'].str.contains(r'BENCHMARK \d+', regex=True) | 
    df_filtered['race_name'].str.contains(r'Class \d+', regex=True)
]

# Extracting "BENCHMARK X" or "CLASS X" and storing it in a new column named 'grade'
BnC['grade'] = BnC['race_name'].str.extract(r'(BENCHMARK \d+|Class \d+)', expand=False)
```

```python
import pandas as pd

# Assuming BnC is your DataFrame that contains the race data

# Fill NaN values in 'track_condition' with an empty string and ensure it's of string type
BnC['track_condition'] = BnC['track_condition'].fillna('').astype(str)

# Filter for winners only (final_rank == 1) and exclude 'UPGRADE' from track_condition
winners = BnC[(BnC['final_rank'] == 1) & (~BnC['track_condition'].str.contains("UPGRADE"))]

# Convert finish_time to timedelta for calculations
winners['cond'] = winners['track_condition']
winners['horse_finish_time'] = pd.to_timedelta(winners['horse_finish_time'])
winners['last_200_section_time'] = pd.to_timedelta(winners['last_200_section_time'])
winners['last_400_section_time'] = pd.to_timedelta(winners['last_400_section_time'])
winners['last_600_section_time'] = pd.to_timedelta(winners['last_600_section_time'])
winners['last_600_total_time'] = pd.to_timedelta(winners['last_600_total_time'])

winners['last_600m'] = winners['last_200_section_time'] + winners['last_400_section_time'] + winners['last_600_section_time'] 



# Group by distance, grade, and track condition to calculate average finish time and race count
standard_times = winners.groupby(['total_distance', 'grade', 'cond', 'course_name']).agg(
    race_count=('horse_finish_time', 'size'),  # Count of races
    finish=('horse_finish_time', 'mean'),
    last_600m=('last_600m', 'mean'),
    _600time=('last_600_section_time', 'mean'),
    _400time=('last_400_section_time', 'mean'),
    _200time=('last_200_section_time', 'mean'),
).reset_index()

# Format the average finish time to show only seconds and milliseconds
standard_times['finish'] = standard_times['finish'].dt.total_seconds()
standard_times['finish'] = standard_times['finish'].apply(lambda x: f"{int(x // 60)}:{x % 60:.2f}")


# Format the average finish time to show only seconds and milliseconds
standard_times['_200time'] = standard_times['_200time'].dt.total_seconds()
standard_times['_200time'] = standard_times['_200time'].apply(lambda x: f"{int(x // 60)}:{x % 60:.2f}")

# Format the average finish time to show only seconds and milliseconds
standard_times['_400time'] = standard_times['_400time'].dt.total_seconds()
standard_times['_400time'] = standard_times['_400time'].apply(lambda x: f"{int(x // 60)}:{x % 60:.2f}")

# Format the average finish time to show only seconds and milliseconds
standard_times['_600time'] = standard_times['_600time'].dt.total_seconds()
standard_times['_600time'] = standard_times['_600time'].apply(lambda x: f"{int(x // 60)}:{x % 60:.2f}")

# Format the average finish time to show only seconds and milliseconds
standard_times['last_600m'] = standard_times['last_600m'].dt.total_seconds()
standard_times['last_600m'] = standard_times['last_600m'].apply(lambda x: f"{int(x // 60)}:{x % 60:.2f}")



# Set display options for nicer output
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', 1000)  # Set desired display width

# Print the calculated standard times
print(standard_times.to_string(index=False, max_colwidth=20))  # Adjust max_colwidth for wrapping

# Reset options back to default (optional)
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
pd.reset_option('display.width')
```
